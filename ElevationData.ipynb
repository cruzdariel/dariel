```python
import requests
import pandas as pd

count = 0

def get_elevation(lat, long, api_key):
    global count
    url = "https://maps.googleapis.com/maps/api/elevation/json"
    params = {
        "locations": f"{lat},{long}",
        "key": api_key
    }

    response = requests.get(url, params=params)
    if response.status_code == 200:
        result = response.json()
        if result["status"] == "OK":
            elevation = result["results"][0]["elevation"]
            count = count + 1
            print(f"Success {count}: Elevation for {lat}, {long} is {elevation} meters")
            return elevation
        else:
            count = count + 1
            print(f"Error {count}: No elevation data found for {lat}, {long}")
            return None
    else:
        count = count + 1
        print(f"Request failed {count}: for {lat}, {long}")
        return None

def add_elevation_to_dataframe(df, api_key):
    df['elevation'] = df.apply(lambda row: get_elevation(row['decimalLatitude'], row['decimalLongitude'], api_key), axis=1)
    return df

df = pd.read_csv("mosquito.csv", low_memory=False)

api_key = 'AIzaSyCh6l1ySrjNGUM8P7gyxfLlzx6V9CJZiUU'
df_with_elevation = add_elevation_to_dataframe(df, api_key)
df_with_elevation.head()
df_with_elevation.to_csv("mosquito.csv")
```
